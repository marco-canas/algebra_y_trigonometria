El documento usa texto en una l√≠nea (sin LaTeX), con secciones y puntos asignados. Puedes ajustar f√°cilmente cualquier enunciado.




```python
# -*- coding: utf-8 -*-
"""
Genera el 'Parcial 1 ‚Äì √Ålgebra y Trigonometr√≠a' en formato .docx
Requiere: pip install python-docx
Las expresiones se escriben en una l√≠nea y sin LaTeX.
"""

from docx import Document
from docx.shared import Pt, Inches
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.oxml.ns import qn
from docx.enum.style import WD_STYLE_TYPE

def add_heading(doc, text, level=0):
    h = doc.add_heading(text, level=level)
    return h

def add_para(doc, text, bold=False, italic=False):
    p = doc.add_paragraph()
    run = p.add_run(text)
    run.bold = bold
    run.italic = italic
    return p

def add_spacer(doc, lines=1):
    for _ in range(lines):
        doc.add_paragraph("")

def set_base_styles(doc):
    # Fuente base
    style = doc.styles['Normal']
    style.font.name = 'Calibri'
    style._element.rPr.rFonts.set(qn('w:eastAsia'), 'Calibri')
    style.font.size = Pt(11)

    # Estilo para enunciados de √≠tem
    if 'Item' not in [s.name for s in doc.styles]:
        item_style = doc.styles.add_style('Item', WD_STYLE_TYPE.PARAGRAPH)
        item_style.font.name = 'Calibri'
        item_style.font.size = Pt(12)
        item_style.font.bold = True

    # Estilo para puntos y sub√≠tems
    if 'Enunciado' not in [s.name for s in doc.styles]:
        en_style = doc.styles.add_style('Enunciado', WD_STYLE_TYPE.PARAGRAPH)
        en_style.font.name = 'Calibri'
        en_style.font.size = Pt(11)

def build_parcial_doc(filename="Parcial_Algebra_Trig_Parcial1_IMUA.docx"):
    doc = Document()
    set_base_styles(doc)

    # Encabezado institucional
    title = doc.add_paragraph()
    title.alignment = WD_ALIGN_PARAGRAPH.CENTER
    r = title.add_run("Instituto de Matem√°ticas - Universidad de Antioquia (IMUA)")
    r.bold = True
    r.font.size = Pt(14)

    add_para(doc, "Parcial 1 ‚Äì √Ålgebra y Trigonometr√≠a", bold=True)
    add_para(doc, "Duraci√≥n: 2 horas")
    add_para(doc, "Puntaje total: 100 (20 puntos por √≠tem).")
    add_para(doc, "Nota: Se evaluar√° claridad, orden l√≥gico, justificaci√≥n de pasos y uso correcto de la notaci√≥n matem√°tica.")
    add_spacer(doc)

    # √çtem 1
    add_para(doc, "1. Teor√≠a y Definiciones (20 pts)", bold=True)
    add_para(doc, "Responda de manera breve y clara:", italic=True)
    add_para(doc, "a) (5 pts) ¬øQu√© se entiende por n√∫mero real? D√© un ejemplo de cada subconjunto: enteros, racionales e irracionales.")
    add_para(doc, "b) (5 pts) Enuncie dos propiedades de los exponentes y dos propiedades de los radicales (incluya condiciones).")
    add_para(doc, "c) (5 pts) ¬øQu√© son los productos notables y por qu√© facilitan los c√°lculos algebraicos?")
    add_para(doc, "d) (5 pts) ¬øQu√© es una expresi√≥n fraccionaria algebraica? D√© un ejemplo.")
    add_spacer(doc)

    # √çtem 2
    add_para(doc, "2. Operaciones con n√∫meros reales y radicales (20 pts)", bold=True)
    add_para(doc, "Resuelva mostrando todos los pasos:", italic=True)
    add_para(doc, "a) (10 pts) Simplifique: (ra√≠z c√∫bica de (16*x^4*y^7)) / (ra√≠z c√∫bica de (2*x*y)).")
    add_para(doc, "b) (10 pts) Simplifique: sqrt(50) - sqrt(18) + sqrt(8).")
    add_spacer(doc)

    # √çtem 3
    add_para(doc, "3. Productos notables y factorizaci√≥n (20 pts)", bold=True)
    add_para(doc, "a) (10 pts) Desarrolle completamente: (x + 3*y)^2.")
    add_para(doc, "b) (10 pts) Factorice: x^2 - 9*x + 20.")
    add_spacer(doc)

    # √çtem 4
    add_para(doc, "4. Expresiones fraccionarias (20 pts)", bold=True)
    add_para(doc, "Simplifique mostrando los pasos:", italic=True)
    add_para(doc, "a) (10 pts) ((x^2 - 4) / (x^2 - 2*x)) * ((x - 2) / (x + 2)).")
    add_para(doc, "b) (10 pts) (3/x) + (2/x^2).")
    add_spacer(doc)

    # √çtem 5
    add_para(doc, "5. Problema de aplicaci√≥n (20 pts)", bold=True)
    add_para(doc, "Un rect√°ngulo tiene largo = 2*x + 4 y ancho = x + 2.")
    add_para(doc, "a) (10 pts) Exprese el per√≠metro y el √°rea en funci√≥n de x.")
    add_para(doc, "b) (10 pts) Si x = sqrt(9), calcule el per√≠metro y el √°rea. Explique las propiedades usadas.")
    add_spacer(doc)

    # √çtem 6 (opcional)
    add_para(doc, "6. (Opcional) Geometr√≠a y factorizaci√≥n ‚Äì Diferencia de cubos (hasta 10 pts)", bold=True)
    add_para(doc, "Considere dos cubos de aristas x y y con x > y.")
    add_para(doc, "a) Calcule los vol√∫menes: x^3 y y^3.")
    add_para(doc, "b) Exprese x^3 - y^3 y factor√≠celo como (x - y)(x^2 + x*y + y^2).")
    add_spacer(doc)

    # R√∫brica (resumen pegado al final para el docente)
    add_para(doc, "R√∫brica de evaluaci√≥n (resumen)", bold=True)
    add_para(doc, "- √çtem 1: Exactitud conceptual (8), Propiedades (6), Ejemplos (4), Claridad (2).")
    add_para(doc, "- √çtem 2: Propiedades (8), Procedimiento (6), Resultado simplificado (4), Presentaci√≥n (2).")
    add_para(doc, "- √çtem 3: Desarrollo (8), Factorizaci√≥n (8), Verificaci√≥n (2), Notaci√≥n (2).")
    add_para(doc, "- √çtem 4: Reducci√≥n por factores (8), Operaciones (6), Resultado (4), Claridad (2).")
    add_para(doc, "- √çtem 5: Modelado (8), C√°lculo num√©rico (6), Justificaci√≥n (4), Presentaci√≥n (2).")
    add_para(doc, "- √çtem 6 (opcional): Vol√∫menes (4), Deducci√≥n y factorizaci√≥n (4), Claridad (2).")

    doc.save(filename)
    return filename

if __name__ == "__main__":
    fname = build_parcial_doc()
    print(f"Documento generado: {fname}")

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np
from matplotlib.patches import Patch

def draw_cuboid(ax, origin, size, color='C0', alpha=0.4, edgecolor='k', linewidth=1.2):
    """Dibuja un paralelep√≠pedo (cuboid) en 3D dado origen (ox,oy,oz) y tama√±o (dx,dy,dz)."""
    ox, oy, oz = origin
    dx, dy, dz = size
    # v√©rtices
    v = np.array([
        [ox,     oy,     oz    ],
        [ox+dx,  oy,     oz    ],
        [ox+dx,  oy+dy,  oz    ],
        [ox,     oy+dy,  oz    ],
        [ox,     oy,     oz+dz ],
        [ox+dx,  oy,     oz+dz ],
        [ox+dx,  oy+dy,  oz+dz ],
        [ox,     oy+dy,  oz+dz ],
    ])
    # caras (listas de v√©rtices)
    faces = [
        [v[j] for j in [0,1,2,3]],  # fondo
        [v[j] for j in [4,5,6,7]],  # techo
        [v[j] for j in [0,1,5,4]],  # frente
        [v[j] for j in [2,3,7,6]],  # atr√°s
        [v[j] for j in [1,2,6,5]],  # derecha
        [v[j] for j in [0,3,7,4]],  # izquierda
    ]
    poly = Poly3DCollection(faces, facecolors=color, linewidths=linewidth, edgecolors=edgecolor, alpha=alpha)
    ax.add_collection3d(poly)
    return v  # devolvemos v√©rtices si hace falta

def set_axes_equal(ax):
    """Asegura proporci√≥n igual en un eje 3D (no nativo en matplotlib)."""
    x_limits = ax.get_xlim3d()
    y_limits = ax.get_ylim3d()
    z_limits = ax.get_zlim3d()
    x_range = abs(x_limits[1] - x_limits[0])
    x_middle = np.mean(x_limits)
    y_range = abs(y_limits[1] - y_limits[0])
    y_middle = np.mean(y_limits)
    z_range = abs(z_limits[1] - z_limits[0])
    z_middle = np.mean(z_limits)
    plot_radius = 0.5*max([x_range, y_range, z_range])
    ax.set_xlim3d([x_middle - plot_radius, x_middle + plot_radius])
    ax.set_ylim3d([y_middle - plot_radius, y_middle + plot_radius])
    ax.set_zlim3d([z_middle - plot_radius, z_middle + plot_radius])

def plot_difference_of_cubes(x=6, y=3, show_labels=True):
    if not (x > y):
        raise ValueError("Se requiere x > y para esta representaci√≥n geom√©trica.")
    fig = plt.figure(figsize=(10,9))
    ax = fig.add_subplot(111, projection='3d')

    # --- Cubo grande (solo aristas) ---
    # Dibujar el cubo de lado x como contorno (sin relleno, para enmarcar)
    draw_cuboid(ax, (0,0,0), (x,x,x), color=(0,0,0,0), alpha=0, edgecolor='k', linewidth=2.0)

    # --- Cubo peque√±o y su etiqueta ---
    draw_cuboid(ax, (0,0,0), (y,y,y), color='green', alpha=0.6, edgecolor='k', linewidth=1.2)
    # centro del cubo peque√±o
    cx_sm, cy_sm, cz_sm = y/2, y/2, y/2
    ax.text(cx_sm, cy_sm, cz_sm, r"$y^3$", color='white', fontsize=12, ha='center', va='center', weight='bold')

    # --- Tres bloques que conforman x^3 - y^3 ---
    # 1) Bloque A: x en [y,x], y in [0,x], z in [0,x]. dims: (x-y, x, x)
    A_origin = (y, 0, 0)
    A_size = (x - y, x, x)
    draw_cuboid(ax, A_origin, A_size, color='tab:blue', alpha=0.35, edgecolor='k', linewidth=1.0)
    A_center = (A_origin[0] + A_size[0]/2, A_origin[1] + A_size[1]/2, A_origin[2] + A_size[2]/2)

    # 2) Bloque B: x in [0,y], y in [y,x], z in [0,x]. dims: (y, x-y, x)
    B_origin = (0, y, 0)
    B_size = (y, x - y, x)
    draw_cuboid(ax, B_origin, B_size, color='tab:blue', alpha=0.35, edgecolor='k', linewidth=1.0)
    B_center = (B_origin[0] + B_size[0]/2, B_origin[1] + B_size[1]/2, B_origin[2] + B_size[2]/2)

    # 3) Bloque C: x in [0,y], y in [0,y], z in [y,x]. dims: (y, y, x-y)
    C_origin = (0, 0, y)
    C_size = (y, y, x - y)
    draw_cuboid(ax, C_origin, C_size, color='tab:blue', alpha=0.35, edgecolor='k', linewidth=1.0)
    C_center = (C_origin[0] + C_size[0]/2, C_origin[1] + C_size[1]/2, C_origin[2] + C_size[2]/2)

    # Etiquetas de vol√∫menes en el interior (contraste seg√∫n color)
    ax.text(A_center[0], A_center[1], A_center[2], r"$(x-y)x^2$", color='black', fontsize=10, ha='center', va='center')
    ax.text(B_center[0], B_center[1], B_center[2], r"$(x-y)xy$", color='black', fontsize=10, ha='center', va='center')
    ax.text(C_center[0], C_center[1], C_center[2], r"$(x-y)y^2$", color='black', fontsize=10, ha='center', va='center')

    # --- L√≠neas de cota (dimensiones x, y, z) ---
    d = 0.12 * x  # desplazamiento visual para las cotas
    # l√≠nea para x (en z = -d, y = -d)
    ax.plot([0, x], [-d, -d], [-d, -d], color='k', linewidth=1.5)
    ax.plot([0, 0], [-d, -d], [-d*0.9, -d*1.1], color='k', linewidth=1.2)  # ticks
    ax.plot([x, x], [-d, -d], [-d*0.9, -d*1.1], color='k', linewidth=1.2)
    ax.text(x/2, -d, -d, r"$x$", fontsize=12, ha='center', va='center')

    # l√≠nea para y (en x = -d, z = -d)
    ax.plot([-d, -d], [0, x], [-d, -d], color='k', linewidth=1.5)
    ax.plot([-d*0.9, -d*1.1], [0, 0], [-d, -d], color='k', linewidth=1.2)
    ax.plot([-d*0.9, -d*1.1], [x, x], [-d, -d], color='k', linewidth=1.2)
    ax.text(-d, x/2, -d, r"$x$", fontsize=12, ha='center', va='center')  # mostrar x en la cota global

    # l√≠nea para z (en x = -d, y = -d)
    ax.plot([-d, -d], [-d, -d], [0, x], color='k', linewidth=1.5)
    ax.plot([-d*0.9, -d*1.1], [-d, -d], [0, 0], color='k', linewidth=1.2)
    ax.plot([-d*0.9, -d*1.1], [-d, -d], [x, x], color='k', linewidth=1.2)
    ax.text(-d, -d, x/2, r"$x$", fontsize=12, ha='center', va='center')

    # cota para y (se√±alada en el eje x en el frente): desde 0 hasta y (en y = -1.7*d to make visible)
    ax.plot([0, y], [-1.6*d, -1.6*d], [0, 0], color='tab:gray', linewidth=2.0)
    ax.text(y/2, -1.6*d, 0, r"$y$", fontsize=12, ha='center', va='center')

    # --- Est√©tica y leyenda ---
    proxies = [
        Patch(facecolor='green', edgecolor='k', label=r"$y^3$"),
        Patch(facecolor='tab:blue', edgecolor='k', label=r"Bloques: $(x-y)x^2,\ (x-y)xy,\ (x-y)y^2$"),
        Patch(facecolor='none', edgecolor='k', label=r"Cubo marco: $x^3$"),
    ]
    ax.legend(handles=proxies, loc='upper left', bbox_to_anchor=(0.02, 0.98))

    # l√≠mites y etiquetas
    ax.set_xlim(-d, x + 0.6*x)
    ax.set_ylim(-d, x + 0.6*x)
    ax.set_zlim(-d, x + 0.6*x)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.view_init(elev=22, azim=-60)

    set_axes_equal(ax)
    plt.title("Descomposici√≥n geom√©trica de $x^3 - y^3$ en $y^3$ y los tres bloques restantes\n(Condici√≥n: x > y)")
    plt.savefig("diferencia_de_cubos.png", dpi=300)
    plt.show()

# Ejecutar con valores por defecto (x > y)
plot_difference_of_cubes(x=6, y=3)
```

## Descripci√≥n geom√©trica (10 min)

1. Dibuja (o muestra) el **cubo grande** de lado $x$. Ese cubo tiene volumen $x^3$.
2. Marca en la esquina inferior el **cubo peque√±o** de lado $y$ (misma orientaci√≥n). Ese cubo tiene volumen $y^3$.
3. Observa que al quitar el cubo peque√±o del grande, queda un ‚Äúrecorte‚Äù que se puede **descomponer en 3 bloques rectangulares** (no solapantes) que, sumados, completan la diferencia de vol√∫menes $x^3 - y^3$.

   * Bloque A: el bloque que queda a lo largo del eje $x$ cuando cortas en $x=y$.
   * Bloque B: el bloque que queda a lo largo del eje $y$ cuando cortas en $y=y$.
   * Bloque C: el bloque que queda sobre el eje $z$ cuando cortas en $z=y$.

(Usa la figura 3D: colorea el cubo $y^3$ y los tres bloques restantes con colores distintos para que se vean separados.)

## C√°lculo de vol√∫menes de las piezas (10‚Äì15 min)

Explica cada bloque en t√©rminos de dimensiones y volumen:

* **Cubo peque√±o**: aristas $y,y,y$ ‚Üí volumen $V_{\text{peque√±o}} = y^3$.

* **Bloque A** (colocado junto al eje $x$, origen en $(y,0,0)$ en la figura):

  * Dimensiones: $(x-y)$ en direcci√≥n $x$, $x$ en direcci√≥n $y$, $x$ en direcci√≥n $z$.
  * Volumen: $V_A = (x-y)\cdot x \cdot x = (x-y)x^2$.

* **Bloque B** (junto al eje $y$, origen en $(0,y,0)$):

  * Dimensiones: $y$ en $x$, $(x-y)$ en $y$, $x$ en $z$.
  * Volumen: $V_B = y\cdot (x-y)\cdot x = (x-y)xy$.

* **Bloque C** (eje $z$, origen en $(0,0,y)$):

  * Dimensiones: $y$ en $x$, $y$ en $y$, $(x-y)$ en $z$.
  * Volumen: $V_C = y\cdot y \cdot (x-y) = (x-y)y^2$.

**Observaci√≥n importante (geometr√≠a):** las tres piezas $A,B,C$ son disjuntas (no se solapan) y junto al cubo peque√±o ocupan exactamente el volumen del cubo grande.

---

## Suma y factor com√∫n (5 min)

Suma las tres piezas que representan la diferencia de vol√∫menes:

$$
x^3 - y^3 = V_A + V_B + V_C
= (x-y)x^2 + (x-y)xy + (x-y)y^2.
$$

Factor com√∫n $(x-y)$:

$$
x^3 - y^3 = (x-y)\bigl(x^2 + xy + y^2\bigr).
$$

Caja final: $\boxed{x^3 - y^3 = (x-y)(x^2+xy+y^2)}$. Esto demuestra la identidad mediante la descomposici√≥n geom√©trica.

## Ejemplo num√©rico (5 min)

Toma $x=6$, $y=3$ (como en el gr√°fico):

* $x^3 = 216$, $y^3 = 27$ ‚Üí diferencia $216-27=189$.
* Vol√∫menes de bloques:

  * $V_A=(6-3)6^2 = 3\cdot36 = 108$.
  * $V_B=(6-3)\cdot6\cdot3 = 3\cdot18 = 54$.
  * $V_C=(6-3)\cdot3^2 = 3\cdot9 = 27$.
* Suma: $108+54+27 = 189$ (coincide).
* Factorizaci√≥n: $(6-3)(36+18+9) = 3\cdot63 = 189$.
