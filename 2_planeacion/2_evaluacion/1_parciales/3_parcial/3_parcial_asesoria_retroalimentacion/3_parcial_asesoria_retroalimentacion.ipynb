Centro: $(2,3)$

# Código con matplotlib


```python
import numpy as np
import matplotlib.pyplot as plt

# Centro y radio
h, k = 2, 3
r = 3

# Parámetro
theta = np.linspace(0, 2*np.pi, 400)

# Ecuación paramétrica
x = h + r * np.cos(theta)
y = k + r * np.sin(theta)

# Gráfico
plt.figure(figsize=(6,6))
plt.plot(x, y, color = 'skyblue')
plt.scatter([h], [k], label="Centro", marker="o")

# Ejes y aspecto
plt.gca().set_aspect('equal')
plt.axhline(0, linewidth=0.5, color = 'green')
plt.axvline(0, linewidth=0.5, color = 'green')

#intersectos
plt.scatter([2, 0, 0], [0, 3 - 5**(1/2), 3 + 5**(1/2)], color = 'red')


plt.title("Circunferencia $(x-2)^2 + (y-3)^2 = 9$")
plt.grid(True)
plt.legend()
plt.show()
```

# Punto 

Si tres vértices consecutivos de un paralelogramo son $A(-1, -3)$, $B(4,2)$ y $C(-7,5)$, encuentre el cuarto vértice 




Aquí tienes **la solución paso a paso** usando únicamente **pendientes** y **distancias euclidianas**, junto con **versiones en LaTeX**, **Python**, **SymPy**, **Matplotlib**, **Plotly** e **ipywidgets**.


El objetivo: **encontrar el cuarto vértice (D)** del paralelogramo (ABCD) cuyos vértices consecutivos son:

$$
A(-1,-3),\quad B(4,2),\quad C(-7,5)
$$

---



# 1. Solución geométrica (solo con pendiente y distancia)

En un **paralelogramo**, los lados opuestos son **paralelos (pendiente igual)** y **congruentes (igual distancia)**.

Vertices consecutivos:
$$
A \rightarrow B \rightarrow C \rightarrow D
$$

Por lo tanto:

* El vector $\overrightarrow{AB}$ debe ser igual al vector $\overrightarrow{CD}$.
* El vector $\overrightarrow{BC}$ debe ser igual al vector $\overrightarrow{AD}$.



### 1.1. Calculemos el vector (\overrightarrow{AB}):

$$
\overrightarrow{AB} = (4 - (-1),\ 2 - (-3)) = (5,5)
$$



### 1.2. Para que (\overrightarrow{CD} = \overrightarrow{AB}):

$$
D = C + \overrightarrow{AB}
$$
$$
D = (-7, 5) + (5, 5) = (-2, 10)
$$



### 1.3. Verificación con pendiente y distancia

Pendientes:

$$
m_{AB} = \frac{5-2}{-1-4} = \frac{5}{5} = 1
$$
$$
m_{CD} = \frac{10-5}{-2 - (-7)} = \frac{5}{5} = 1
$$



Distancias:

$$
AB = \sqrt{5^2 + 5^2} = \sqrt{50}
$$
$$
CD = \sqrt{5^2 + 5^2} = \sqrt{50}
$$

Los lados son paralelos y congruentes → **Correcto**.

---



# **Cuarto vértice del paralelogramo**

$$
\boxed{D(-2,\ 10)}
$$

---



# 2. Código en Python (básico)


```python
A = (-1, -3)
B = (4, 2)
C = (-7, 5)

AB = (B[0] - A[0], B[1] - A[1])
D = (C[0] + AB[0], C[1] + AB[1])

D

```


#  3. Código con SymPy



```python
import sympy as sp

A = sp.Point(-1, -3)
B = sp.Point(4, 2)
C = sp.Point(-7, 5)

D = C + (B - A)

D
```


# 4. Visualización con Matplotlib


```python
import matplotlib.pyplot as plt

A = (-1, -3)
B = (4, 2)
C = (-7, 5)
D = (-2, 10)

X = [A[0], B[0], C[0], D[0], A[0]]
Y = [A[1], B[1], C[1], D[1], A[1]]

plt.figure(figsize=(6,6))
plt.plot(X, Y, marker='o')
plt.text(*A, "A")
plt.text(*B, "B")
plt.text(*C, "C")
plt.text(*D, "D")

plt.grid(True)
plt.gca().set_aspect('equal')
plt.title("Paralelogramo ABCD")
plt.show()
```

# 5. Visualización con Plotly


```python
import plotly.graph_objects as go

A = (-1, -3)
B = (4, 2)
C = (-7, 5)
D = (-2, 10)

X = [A[0], B[0], C[0], D[0], A[0]]
Y = [A[1], B[1], C[1], D[1], A[1]]

fig = go.Figure()

fig.add_trace(go.Scatter(x=X, y=Y, mode='lines+markers+text',
                         text=['A','B','C','D','A'],
                         textposition="top center"))

fig.update_layout(title="Paralelogramo ABCD",
                  xaxis=dict(scaleanchor="y"))

fig.show()
```

# 6. Visualización interactiva con ipywidgets

```python
import matplotlib.pyplot as plt
from ipywidgets import interact

A = (-1, -3)
B = (4, 2)
C = (-7, 5)

def plot_parallelogram(scale=1.0):
    AB = ((B[0]-A[0])*scale, (B[1]-A[1])*scale)
    D = (C[0] + AB[0], C[1] + AB[1])

    X = [A[0], B[0], C[0], D[0], A[0]]
    Y = [A[1], B[1], C[1], D[1], A[1]]

    plt.figure(figsize=(6,6))
    plt.plot(X, Y, marker='o')
    plt.grid(True)
    plt.gca().set_aspect('equal')
    plt.title("Paralelogramo Escalable ABCD")
    plt.show()

interact(plot_parallelogram, scale=(0.5, 2.0, 0.1))
```
