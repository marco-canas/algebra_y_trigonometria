# Prompt para convertir a formato .docx  

Dame el script de Python para convertir en archivo .docx a este quiz, sin utilizar latex, y utilizando símbolos unicode para hacer legible mas expresiones matemáticas, por favor, presenta las expresiones fraccionarias con notación a dos líneas para facilitar su legibilidad.  


```python
from docx import Document
from docx.shared import Pt

# Crear documento
doc = Document()

# Título
doc.add_heading("Quiz – Unidad 2: Ecuaciones e Inecuaciones", level=1)

# Lista de preguntas
preguntas = [
    "1. Ecuaciones lineales\n\nResuelva la ecuación lineal:\n\n7x − 5 = 16",
    
    "2. Relación entre escalas de temperatura\n\n"
    "La relación entre Celsius (C) y Fahrenheit (F) está dada por:\n\n"
    "   C = 5/9 × (F − 32)\n\n"
    "Despeje F en términos de C.",
    
    "3. Ecuaciones cuadráticas\n\nResuelva la ecuación cuadrática:\n\n2x² − 3x − 5 = 0",
    
    "4. Método de completación de trinomio cuadrado perfecto\n\n"
    "Deduzca la fórmula general para resolver la ecuación cuadrática:\n\n"
    "   x² + bx + c = 0\n\n"
    "utilizando el método de completación del trinomio cuadrado perfecto.",
    
    "5. Ecuaciones racionales\n\nResuelva la ecuación:\n\n"
    "   (2x − 3)\n   -------  = 0\n   (x + 1)",
    
    "6. Mezcla de productos químicos\n\n"
    "Un químico tiene 10 ml de una solución que contiene ácido al 30%.\n"
    "¿Cuántos mililitros de ácido puro deben agregarse para que la\n"
    "solución final tenga una concentración al 50%?",
    
    "7. Inecuaciones lineales\n\nResuelva la inecuación lineal:\n\n4x − 7 < 5",
    
    "8. Inecuaciones racionales\n\nResuelva la inecuación:\n\n"
    "   (x − 2)\n   -------  < 0\n   (x + 3)"
]

# Fuente uniforme
style = doc.styles['Normal']
font = style.font
font.name = 'Calibri'
font.size = Pt(12)

# Agregar preguntas
for pregunta in preguntas:
    doc.add_paragraph(pregunta)
    doc.add_paragraph()  # espacio

# Guardar archivo
doc.save("Quiz3_1_Unidad2.docx")
print("Archivo 'Quiz_Unidad2.docx' creado exitosamente.")
```
