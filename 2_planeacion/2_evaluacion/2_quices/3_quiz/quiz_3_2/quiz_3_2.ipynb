# üìò Gu√≠a Did√°ctica

## Problema

> Un qu√≠mico tiene 1 litro de una soluci√≥n que contiene √°cido al 10%.
> ¬øCu√°ntos mililitros de √°cido puro deben agregarse para que la concentraci√≥n aumente al 60%?

---

## ‚úçÔ∏è Planteamiento matem√°tico (en LaTeX)

```latex
\section*{Planteamiento del problema}

Tenemos una soluci√≥n de $1$ litro (equivalente a $1000$ mL) que contiene √°cido al $10\%$.

\begin{itemize}
    \item Volumen inicial de la soluci√≥n: $1000$ mL.
    \item Concentraci√≥n inicial: $10\%$.
    \item √Åcido presente inicialmente: 
    \[
        0.10 \times 1000 = 100 \, \text{mL}.
    \]
    \item Sea $x$ la cantidad (en mL) de √°cido puro que se agrega.
\end{itemize}

Despu√©s de agregar $x$ mililitros de √°cido puro:

\[
\text{√Åcido total} = 100 + x
\]
\[
\text{Volumen total de la soluci√≥n} = 1000 + x
\]

La nueva concentraci√≥n debe ser del $60\%$:

\[
\frac{100 + x}{1000 + x} = 0.60
\]

---
\section*{Resolviendo la ecuaci√≥n}
Multiplicamos en cruz:

\[
100 + x = 0.60 \cdot (1000 + x)
\]

\[
100 + x = 600 + 0.60x
\]

\[
x - 0.60x = 600 - 100
\]

\[
0.40x = 500
\]

\[
x = \frac{500}{0.40} = 1250
\]

---
\section*{Respuesta}
Se deben agregar \textbf{1250 mL} de √°cido puro.
```

## üêç Soluci√≥n en Python

```python
# ===========================================
# Problema: Mezcla de soluciones qu√≠micas
# ===========================================

# Volumen inicial en mL
V_inicial = 1000  

# √Åcido puro inicial
A_inicial = 0.10 * V_inicial  

# Cantidad a agregar (x)
# Ecuaci√≥n: (A_inicial + x) / (V_inicial + x) = 0.60

import sympy as sp

x = sp.symbols('x', positive=True)
ecuacion = sp.Eq((A_inicial + x) / (V_inicial + x), 0.60)

# Resolver
solucion = sp.solve(ecuacion, x)[0]
print(f"Se deben agregar {solucion} mL de √°cido puro.")
```
