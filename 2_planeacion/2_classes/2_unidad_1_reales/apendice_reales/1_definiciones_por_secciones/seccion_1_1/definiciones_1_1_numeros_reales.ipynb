# generación del QR para el cuaderno de esta clase    

Aquí tienes una función de Python que genera un código QR a partir de una URL y lo muestra en un cuaderno Jupyter:


````python
import qrcode
from IPython.display import display

def generar_qr_notebook(url):
    """
    Genera un código QR para una URL y lo muestra en un cuaderno Jupyter.
    
    Parámetros:
    url (str): La URL para la cual se generará el código QR.
    
    Retorna:
    None (pero muestra el código QR en el notebook)
    """
    # Crear el objeto QR
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    
    # Añadir la URL al código QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color="black", back_color="white")
    
    # Mostrar la imagen en el notebook
    display(img)

````

# 1. Título de la clase y definiciones (Sección 1.1 — *Números reales*)

**Título de la clase:** *Definiciones y estructura del conjunto de los Números Reales — Sección 1.1 (Swokowski, 13ª ed.)*



A continuación, **numeradas**, las definiciones clave de la sección 1.1 (formuladas didácticamente, con \$...\$ para LaTeX):



1. Definición de **Números naturales** $ \mathbb{N}$: conjunto de enteros positivos usados para contar $(1,2,3,\dots)$.


2. Definición de **Números enteros** $ \mathbb{Z} $: $ \{ \dots, -2, -1, 0, 1, 2, \dots \}$ — extendiendo $\mathbb{N}$ incluyendo ceros y negativos.


3. Definición de **Números racionales** \$ \mathbb{Q} \$: números que pueden expresarse como fracción \$ \dfrac{p}{q} \$ con \$p\in\mathbb{Z}, q\in\mathbb{Z}\setminus{0}\$.


4. Definición de **Números irracionales**: números que no pueden expresarse como \$ \dfrac{p}{q} \$ (ej.: \$\sqrt{2}, \pi, e\$).


5. Definición de **Números reales** \$ \mathbb{R} \$: unión de \$\mathbb{Q}\$ y el conjunto de irracionales; representan todos los puntos de la recta real.
6. Definición de **Valor absoluto**: \$|x|=\begin{cases} x & x\ge 0\ -x & x<0 \end{cases}\$.
7. Definición de **Orden en \$\mathbb{R}\$**: para \$a,b\in\mathbb{R}\$, decimos \$a\<b\$ si \$b-a\$ es positivo; propiedades de transitividad, antisimetría, etc.
8. Definición de **Intervalos**: abierto \$(a,b)\$, cerrado $\[a,b]\$, semiabiertos \$(a,b]\$, $\[a,b)\$, medios.
9. Definición de **Densidad de \$\mathbb{Q}\$ en \$\mathbb{R}\$**: entre dos reales siempre hay un racional.
10. Definición de **Máximo y mínimo** (cuando existen) para subconjuntos acotados.
11. Definición de **Supremo (sup)** y **Ínfimo (inf)**: menor cota superior y mayor cota inferior, respectivamente.
12. Definición de **Axioma de completitud**: todo subconjunto no vacío acotado superiormente tiene supremo en \$\mathbb{R}\$.
13. Definición de **Decimal finito** y **decimal periódico** (racionales) vs. expansiones decimales no periódicas (irracionales).
14. Definición de **Propiedades algebraicas de \$\mathbb{R}\$**: cerradura bajo suma y multiplicación, existencia de inversos aditivos y multiplicativos (salvo \$0\$), leyes distributiva, asociativa y conmutativa.
15. Definición de **Representación en la recta real**: cada \$x\in\mathbb{R}\$ corresponde a un punto único en la recta real.



*(Estas definiciones están parafraseadas y adaptadas pedagógicamente a partir de la sección indicada en Swokowski; en la clase se hará referencia textual a la sección original para ver redacción exacta.)*

```python
from graphviz import Digraph

dot = Digraph('Mapa_Numeros_Reales', format='png', engine='neato')

dot.node('N', 'Naturales\n$\\mathbb{N}$', shape='triangle', fillcolor='#FFFFCC')
dot.node('sup', 'Supremo / Ínfimo', shape='house', fillcolor='#FFDADA')
dot.node('dec', 'Expansión decimal\n(finitas, periódicas, no periódicas)', shape='note', fillcolor='#E6F2FF')
dot.node('axiom', 'Axioma de\ncompletitud', shape='diamond', fillcolor='#D6FFE6')

# Aristas
dot.edge('N','Z', label='inclusión')
dot.edge('Z','Q', label='inclusión')
dot.edge('Q','R', label='densidad en')
dot.edge('I','R', label='inclusión')
dot.edge('sup','R', label='definido en')
dot.edge('axiom','sup', label='garantiza existencia')
dot.edge('dec','Q', label='racionales => periódicas/fin.')
dot.edge('dec','I', label='irracionales => no periódicas')

# detalles conceptuales
dot.node('abs','Valor absoluto\n$|x|$', shape='circle', fillcolor='#F2E6FF')
dot.edge('R','abs')
dot.edge('abs','sup', style='dashed', label='usar en cotas')

# render
output_path = r"C:\Users\marco\Downloads\mapa_numeros_reales"
dot.render(filename=output_path, view=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# secuencia por biseccion entre 1 y 2
def bisection_steps(a, b, n_steps=20):
    steps = []
    for _ in range(n_steps):
        m = (a + b) / 2
        if m*m < 2:
            a = m
        else:
            b = m
        steps.append((a, b, m))
    return steps

steps = bisection_steps(1,2,30)

fig, ax = plt.subplots(figsize=(8,1.5))
ax.set_xlim(0.8,2.2)
ax.get_yaxis().set_visible(False)
point, = ax.plot([], [], 'o', markersize=8)

def init():
    point.set_data([], [])
    return (point,)

def update(i):
    a,b,m = steps[i]
    point.set_data(m, 0)
    ax.set_title(f"step {i+1}: m={m:.8f}")
    return (point,)

ani = FuncAnimation(fig, update, frames=len(steps), init_func=init, blit=True, interval=300)
# Render in Jupyter
```

```python
def classify(x, tol=1e-10):
```
