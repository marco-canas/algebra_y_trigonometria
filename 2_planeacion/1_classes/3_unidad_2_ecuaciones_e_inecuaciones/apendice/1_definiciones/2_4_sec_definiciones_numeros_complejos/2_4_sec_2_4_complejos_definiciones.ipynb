# generaci칩n del QR para el cuaderno de esta clase    

Aqu칤 tienes una funci칩n de Python que genera un c칩digo QR a partir de una URL y lo muestra en un cuaderno Jupyter:


```python
import qrcode
from IPython.display import display

def generar_qr_notebook(url):
    """
    Genera un c칩digo QR para una URL y lo muestra en un cuaderno Jupyter.
    
    Par치metros:
    url (str): La URL para la cual se generar치 el c칩digo QR.
    
    Retorna:
    None (pero muestra el c칩digo QR en el notebook)
    """
    # Crear el objeto QR
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    
    # A침adir la URL al c칩digo QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color="black", back_color="white")
    
    # Mostrar la imagen en el notebook
    display(img)

```

```python
generar_qr_notebook()
```


### Notas:
