# generación del QR para el cuaderno de esta clase    

Aquí tienes una función de Python que genera un código QR a partir de una URL y lo muestra en un cuaderno Jupyter:


```python
import qrcode
from IPython.display import display

def generar_qr_notebook(url):
    """
    Genera un código QR para una URL y lo muestra en un cuaderno Jupyter.
    
    Parámetros:
    url (str): La URL para la cual se generará el código QR.
    
    Retorna:
    None (pero muestra el código QR en el notebook)
    """
    # Crear el objeto QR
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    
    # Añadir la URL al código QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color="black", back_color="white")
    
    # Mostrar la imagen en el notebook
    display(img)

```

```python
generar_qr_notebook()
```


### Notas:
- La función usa la biblioteca `qrcode` para generar el código y `IPython.display` para mostrarlo en el notebook.
- El código QR aparecerá como salida de la celda donde llames a la función.
- Puedes ajustar parámetros como `box_size` (tamaño de los cuadrados) o `border` (tamaño del borde) según tus necesidades.

## Ley de Boyle   

La ley de Boyle para cierto gas expresa que si la temperatura es constante, entonces $pv = 200$, donde $p$ es la presión (en $lb/in^{2}$) y $v$ es el volumen (en $in^{3}$). Si $25 \leq v \leq 50$, ¿cuál es el correspondiente intervalo para $p$?


## Solución:  

El que nos digan que la relación entre $p$ y $v$ se puede describir como $pv = 200$, nos dice que la relación entre estas dos variables es de proporcionalidad inversa.  

$$ p = \frac{200}{v} $$

$$
25 \le v \le 50
$$  

Aplicando el operador recíproco a toda esta desigualdad continua: (lo que cambia el sentido de la desigualdad)  


$$
\frac{1}{25} \ge \frac{1}{v} \ge \frac{1}{50}
$$

multiplicando por 200 obtenemos: 


$$
\frac{200}{25} \ge \frac{200}{v} \ge \frac{200}{50}
$$

Aplicando definición de la relación de proporcionalidad dada: 

$$ \frac{200}{25} \ge p \ge \frac{200}{50} $$

realizando las divisiones: 

$$ 8 \ge p \ge 4 $$

que es equivelente a 

$$ 4 \le p \le 8 $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición del dominio y la función
v = np.linspace(25, 50, 200)
p = 200 / v

# Creación del gráfico
plt.figure(figsize=(8, 5))
plt.plot(v, p, label=r"$p = \frac{200}{v}$", linewidth=2)
plt.title("Gráfico de $p = \\frac{200}{v}$ en el dominio [25, 50]", fontsize=14)
plt.xlabel("v", fontsize=12)
plt.ylabel("p", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.savefig(r"C:\Users\marco\Downloads\ejercicio_85_ejer_repaso_cap_2.jpg")
plt.show()
```

Este gráfico confirma, que el el volumen está entre 25 y 50, entonces la presión se moverá entre 4 y 8.
