# generación del QR para el cuaderno de esta clase    

Aquí tienes una función de Python que genera un código QR a partir de una URL y lo muestra en un cuaderno Jupyter:


````python
import qrcode
from IPython.display import display

def generar_qr_notebook(url):
    """
    Genera un código QR para una URL y lo muestra en un cuaderno Jupyter.
    
    Parámetros:
    url (str): La URL para la cual se generará el código QR.
    
    Retorna:
    None (pero muestra el código QR en el notebook)
    """
    # Crear el objeto QR
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    
    # Añadir la URL al código QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color="black", back_color="white")
    
    # Mostrar la imagen en el notebook
    display(img)

````

````python
generar_qr_notebook()
````


### Notas:
- La función usa la biblioteca `qrcode` para generar el código y `IPython.display` para mostrarlo en el notebook.
- El código QR aparecerá como salida de la celda donde llames a la función.
- Puedes ajustar parámetros como `box_size` (tamaño de los cuadrados) o `border` (tamaño del borde) según tus necesidades.

## Ejemplo de plantilla de solución (ejercicio 15) — pasos

**(a) Descripción geométrica:**
Grafique $y_1(x)=\sqrt{2x+3}$ y $y_2(x)=x-1$ en la misma ventana. Las intersecciones corresponden a soluciones reales. Restricción: $2x+3 \ge 0 \Rightarrow x \ge -3/2$.

**(b) Descripción verbal / secuencial del algoritmo:**

1. Determinar dominio: $x \ge -1.5$.
2. Elevar ambos miembros al cuadrado (tener cuidado con raíces extraviadas): $(\sqrt{2x+3})^2 = (x-1)^2$.
3. Resolver $2x+3 = x^2 -2x +1$ → formar polinomio, factorizar.
4. Verificar cada solución en la ecuación original (sustituir y comprobar no-negatividad).

**(c) Implementación en Python (Jupyter / Powershell):**

```python
# archivo: resolver_ej15.py  (se puede ejecutar con: python resolver_ej15.py desde PowerShell)
import sympy as sp
x = sp.symbols('x', real=True)
expr = sp.sqrt(2*x+3) - (x-1)
sol = sp.solve(sp.Eq(sp.sqrt(2*x+3), x-1), x)
# Alternativa: resolver el polinomio cuadrático derivado:
sq = sp.solve(sp.Eq(2*x+3, (x-1)**2), x)
valid = [s for s in sq if (s>=-sp.Rational(3,2)) and sp.nsimplify(sp.sqrt(2*s+3)) == sp.nsimplify(s-1)]
print("Soluciones candidatas:", sq)
print("Soluciones válidas (verificadas):", valid)
```

**(d) Cálculo a mano:**
$2x+3 = x^2 -2x +1 \Rightarrow 0 = x^2 -4x -2 \Rightarrow x = 2 \pm \sqrt{4+2} = 2 \pm \sqrt{6}$. Verificar: plug into original; solo las que satisfacen $x-1 \ge 0$ y raíces reales.

```python
RAW_URL = "https://github.com/marco-canas/groups_list/raw/main/vector_geometry_and_linear_algebra.xlsx"

def asignar_tareas_por_ubicacion(raw_url=RAW_URL, hoja=0, seed=2025):
    random.seed(seed)
    df = pd.read_excel(raw_url, sheet_name=hoja)  # requiere internet al correr
    ejercicios = list(range(1,21))  # 20 ejercicios numerados
    asignaciones = []
    for _, row in df.iterrows():
        nombre = row.get('nombre') or row.get('Nombre') or row.get('student') or 'Estudiante'
        ubicacion = row.get('ubicacion') or row.get('ubicación') or row.get('Location') or 'Unknown'
        # Estrategia: asignar ejercicio basado en hash de ubicación + aleatoriedad
        idx = abs(hash(str(ubicacion))) % len(ejercicios)
        # rotación aleatoria por estudiante
        offset = random.randint(0, len(ejercicios)-1)
        ejercicio = ejercicios[(idx + offset) % len(ejercicios)]
        asignaciones.append({'nombre': nombre, 'ubicacion': ubicacion, 'ejercicio': ejercicio})
    out = pd.DataFrame(asignaciones)
    out.to_csv("asignaciones_ejercicios_seccion2_5.csv", index=False)
    return out

if __name__ == "__main__":
    tabla = asignar_tareas_por_ubicacion()
    print(tabla.head())
```

4. Feedback final del docente con observaciones pedagógicas sobre registros semióticos.

Rúbrica rápida: claridad de representación (20%), validez matemática (30%), calidad del código y visualizaciones (25%), verificación y reflexión semiótica (25%).

**Ejemplo: animación de cómo varía la intersección al cambiar parámetro $k$** (usar en Jupyter; `IPython.display.HTML` para renderizar):

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

def animate_intersection(a=1,b=0,c=0,d=0):
    fig, ax = plt.subplots(figsize=(6,4))
    x = np.linspace(-5,5,400)
    k_values = np.linspace(-3,3,120)
    line1, = ax.plot([], [], lw=2)
    line2, = ax.plot([], [], lw=2)
    title = ax.text(0.5,1.05,"", transform=ax.transAxes, ha="center")

    def init():
        ax.set_xlim(-5,5)
        ax.set_ylim(-10,10)
        return line1, line2, title

    def update(i):
        k = k_values[i]
        y1 = a*x**3 + b*x**2 + c*x + d - k
        y2 = np.cbrt(a*x**2 + b)  # ejemplo
        line1.set_data(x, y1)
        line2.set_data(x, y2)
        title.set_text(f"k = {k:.2f}")
        return line1, line2, title

    ani = FuncAnimation(fig, update, frames=len(k_values), init_func=init, blit=True)
    plt.close(fig)
    return HTML(ani.to_jshtml())
```
