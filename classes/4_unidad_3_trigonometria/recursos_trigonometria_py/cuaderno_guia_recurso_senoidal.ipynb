````python
# codigo del recurso de estudio para la funcion senoidal 

import numpy as np # Genera la progresi√≥n aritm√©tica para el dominio de graficaci√≥n. 
import matplotlib.pyplot as plt # Grafica la funci√≥n y = y(x) en el dominio de graficaci√≥n.
from ipywidgets import interact # Interactividad o deslizados para mover los par√°metros A, B, C, y D 
#definir dominio  
a, b = 0, 4*np.pi
# definir conjunto de llegada
c, d = -6, 6
@interact(A=(1, 5), B = (1, 5), C = (0, 2*np.pi, 0.1), D = (-2, 2))   

def graficar_seno(A=1, B=1, C=0, D=0):
    x = np.linspace(0, 4*np.pi, 200) # crea una secuencia de 201 n√∫meros entre 0 y 4œÄ 
    y = A * np.sin(B*x + C) + D # 
    plt.figure(figsize=(8, 3)) # genera un lienzo o figura de 10 x 4 pulgadas 
    plt.title('Exploraci√≥n a las gr√°ficas de $y = A sen(Bx + C) + D$')
    plt.axis([a, b, c, d]) # defines el rectangulo de visualizaci√≥n de la gr√°fica
    plt.axhline(y = 0, color = 'blue')
    plt.axvline(x = 0)
    plt.xticks(np.arange(a,b+np.pi/2, np.pi/2), ['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '6pi/2', '7pi/2', '4pi'] ) # metrizar el eje horizontal o escalar los ejes. 
    plt.yticks(np.arange(c,d+1)) # metrizar el eje vertical 
    plt.plot(x, y, label=f"${A}\\cdot\\sin({B}x + {C:.1f}) + {D}$", color = 'red') # Hace gr√°ficos de l√≠neas 
    plt.legend('Exploraci√≥n de las gr√°ficas de $y = A \cdot \sin(Bx + C) + D$')
    plt.grid()
    plt.savefig(r"C:\Users\marco\Downloads\exploracion_graficas_y_A_sin.jpg")
    plt.show()

````

$$ y = A \cdot \sin(Bx + C) + D $$  
Los estudiantes aprender√°n a diferenciar:

- **Sesgo (D)**: Desplazamiento vertical.

### **3. Explicaci√≥n de los Par√°metros**  
**Demuestra c√≥mo cada par√°metro afecta la gr√°fica:**  

1. **Amplitud (A):**  
   - **Rango:** 1 a 5.  
   - **Efecto:** Controla la altura de la onda.  
   - **Ejemplo:**  
     - `A=1`: Oscila entre -1 y 1.  
     - `A=3`: Oscila entre -3 y 3.  

2. **Periodo (B):**  
   - **Rango:** 1 a 5.  
   - **Efecto:** Determina cu√°nto tarda un ciclo completo.  
     - F√≥rmula: `Periodo = 2œÄ / B`.  
   - **Ejemplo:**  
     - `B=1`: Periodo = 2œÄ.  
     - `B=2`: Periodo = œÄ (ciclos m√°s cortos).  

3. **Fase (C):**  
   - **Rango:** 0 a 2œÄ.  
   - **Efecto:** Desplaza la gr√°fica horizontalmente.  
   - **Ejemplo:**  
     - `C=0`: Sin desplazamiento.  
     - `C=œÄ/2`: Desplazamiento a la izquierda en œÄ/2.  

4. **Sesgo (D):**  
   - **Rango:** -2 a 2.  
   - **Efecto:** Desplaza la gr√°fica verticalmente.  
   - **Ejemplo:**  
     - `D=0`: Centrada en y=0.  
     - `D=1.5`: Sube toda la gr√°fica 1.5 unidades.
