# generaci칩n del QR para el cuaderno de esta clase    

Aqu칤 tienes una funci칩n de Python que genera un c칩digo QR a partir de una URL y lo muestra en un cuaderno Jupyter:


````python
import qrcode
from IPython.display import display

def generar_qr_notebook(url):
    """
    Genera un c칩digo QR para una URL y lo muestra en un cuaderno Jupyter.
    
    Par치metros:
    url (str): La URL para la cual se generar치 el c칩digo QR.
    
    Retorna:
    None (pero muestra el c칩digo QR en el notebook)
    """
    # Crear el objeto QR
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    
    # A침adir la URL al c칩digo QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color="black", back_color="white")
    
    # Mostrar la imagen en el notebook
    display(img)

````

````python
generar_qr_notebook()
````


### Notas:
- La funci칩n usa la biblioteca `qrcode` para generar el c칩digo y `IPython.display` para mostrarlo en el notebook.
- El c칩digo QR aparecer치 como salida de la celda donde llames a la funci칩n.
- Puedes ajustar par치metros como `box_size` (tama침o de los cuadrados) o `border` (tama침o del borde) seg칰n tus necesidades.

# 游닂 Gu칤a Did치ctica: Problema de Facturaci칩n

## 游닇 Enunciado

\*"El tiempo de una ingeniera consultora se factura a \$60 por hora y el de su asistente se factura a \$20 por hora.

Un cliente recibe una cuenta de \$580 por cierto trabajo. Si la asistente trabaj칩 5 horas menos que la ingeniera, 쯖u치nto tiempo factur칩 cada una en el trabajo?"\*

---



## 1. **Planteamiento algebraico**

* Sea $x$ = horas trabajadas por la **ingeniera**.
* Entonces la asistente trabaj칩 $x - 5$ horas.

Costo total:

$$
60x + 20(x - 5) = 580
$$

---



## 2. **Resolviendo la ecuaci칩n**

$$
60x + 20x - 100 = 580
$$



$$
80x = 680
$$



$$
x = 8.5
$$

* Ingeniera: **8.5 h**
* Asistente: $8.5 - 5 = 3.5$ h

---



## 3. **Verificaci칩n**

$$
60(8.5) + 20(3.5) = 510 + 70 = 580 \quad 九
$$

---



## 4. **Representaci칩n con Python**

Aqu칤 tienes un **script en Python** para usarlo en clase con **pandas**, **matplotlib** y **plotly**:


````python

import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

# Datos
ingeniera_horas = 8.5
asistente_horas = 3.5
costo_ingeniera = 60 * ingeniera_horas
costo_asistente = 20 * asistente_horas
total = costo_ingeniera + costo_asistente

# Tabla en pandas
data = {
    "Persona": ["Ingeniera", "Asistente"],
    "Horas trabajadas": [ingeniera_horas, asistente_horas],
    "Tarifa por hora ($)": [60, 20],
    "Costo total ($)": [costo_ingeniera, costo_asistente]
}
df = pd.DataFrame(data)

print("Tabla de facturaci칩n:\n")
print(df)
print("\nTotal facturado:", total)

# --- Visualizaci칩n con Matplotlib ---
plt.bar(df["Persona"], df["Costo total ($)"])
plt.title("Distribuci칩n de la factura entre Ingeniera y Asistente")
plt.ylabel("Costo ($)")
plt.show()

# --- Visualizaci칩n con Plotly ---
fig = px.pie(df, names="Persona", values="Costo total ($)",
             title="Distribuci칩n porcentual de costos")
fig.show()

````


## 5. **Interpretaci칩n de resultados**

* La **ingeniera** trabaj칩 **8.5 horas**, representando el **88% del costo total**.
* La **asistente** trabaj칩 **3.5 horas**, representando el **12% del costo total**.

Esto muestra la diferencia de tarifas y la importancia del tiempo facturado en la distribuci칩n final de la cuenta.
