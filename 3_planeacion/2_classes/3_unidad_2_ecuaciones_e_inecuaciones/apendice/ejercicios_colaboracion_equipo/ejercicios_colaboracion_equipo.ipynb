# generaci√≥n del QR para el cuaderno de esta clase    

Aqu√≠ tienes una funci√≥n de Python que genera un c√≥digo QR a partir de una URL y lo muestra en un cuaderno Jupyter:


````python
import qrcode
from IPython.display import display

def generar_qr_notebook(url):
    """
    Genera un c√≥digo QR para una URL y lo muestra en un cuaderno Jupyter.
    
    Par√°metros:
    url (str): La URL para la cual se generar√° el c√≥digo QR.
    
    Retorna:
    None (pero muestra el c√≥digo QR en el notebook)
    """
    # Crear el objeto QR
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    
    # A√±adir la URL al c√≥digo QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color="black", back_color="white")
    
    # Mostrar la imagen en el notebook
    display(img)

````

````python
generar_qr_notebook()
````


### Notas:
- La funci√≥n usa la biblioteca `qrcode` para generar el c√≥digo y `IPython.display` para mostrarlo en el notebook.
- El c√≥digo QR aparecer√° como salida de la celda donde llames a la funci√≥n.
- Puedes ajustar par√°metros como `box_size` (tama√±o de los cuadrados) o `border` (tama√±o del borde) seg√∫n tus necesidades.

Perfecto üëå Te dejo un script en Python que toma la lista de estudiantes desde el archivo `.xlsx` que ya tienes, reparte de forma equitativa todos los ejercicios del **cap√≠tulo 2 de Swokowski** y genera un archivo nuevo con la asignaci√≥n.

````python
import pandas as pd
import random

# =============================
# 1. Cargar lista de estudiantes
# =============================
file_path = r"C:\Users\marco\Documentos\docencia\groups_list\algebra_and_trigonometry.xlsx"   # <- Ajusta la ruta si es necesario
students = pd.read_excel(file_path, sheet_name="Sheet1")

# =============================
# 2. Definir n√∫mero de ejercicios por secci√≥n
# =============================
exercises_per_section = {
    "Secci√≥n 2.1 (Ecuaciones)": 78,
    "Secci√≥n 2.2 (Problemas de aplicaci√≥n)": 40,
    "Secci√≥n 2.3 (Ecuaciones cuadr√°ticas)": 83,
    "Secci√≥n 2.4 (N√∫meros complejos)": 62,
    "Secci√≥n 2.5 (Otros tipos de ecuaciones)": 71,
    "Secci√≥n 2.6 (Desigualdades)": 89,
    "Secci√≥n 2.7 (Valor absoluto)": 55,
}

# =============================
# 3. Construir lista completa de ejercicios
# =============================
all_exercises = []
for section, count in exercises_per_section.items():
    all_exercises.extend([f"{section} - Ejercicio {i+1}" for i in range(count)])

# Mezclar aleatoriamente los ejercicios
random.shuffle(all_exercises)

# =============================
# 4. Repartir los ejercicios entre estudiantes
# =============================
num_students = len(students)
assignments = {student: [] for student in students["Nombre"]}

for i, exercise in enumerate(all_exercises):
    student = students["Nombre"].iloc[i % num_students]
    assignments[student].append(exercise)

# =============================
# 5. Crear DataFrame con asignaciones
# =============================
assignments_df = pd.DataFrame([
    {"Nombre": student, "Ejercicios asignados": ", ".join(ex_list)}
    for student, ex_list in assignments.items()
])

# =============================
# 6. Guardar en archivo Excel
# =============================
output_path = "asignacion_ejercicios_cap2.xlsx"
assignments_df.to_excel(output_path, index=False)

print(f"‚úÖ Archivo generado: {output_path}")
print(assignments_df.head())

````
