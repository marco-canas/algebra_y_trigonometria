### **1. Teoría y Definiciones (20 puntos)**

Responda brevemente y con precisión:

a) Defina el **conjunto de los números reales**. ¿Qué subconjuntos lo componen? (4 pts)
b) ¿Qué significa que un exponente sea racional? Dé un ejemplo. (4 pts)
c) ¿Qué es un producto notable? Mencione al menos dos con su fórmula. (4 pts)
d) Explique por qué no es posible dividir entre cero en una fracción algebraica. (4 pts)
e) ¿Qué se entiende por “racionalizar” un denominador? (4 pts)

### **2. Resolución de Ecuaciones y Simplificación (20 puntos)**

$$
\frac{\sqrt{18}}{\sqrt{2}}

(5 pts)

b) Racionalice el denominador y simplifique:

$$
\frac{5}{2 + \sqrt{3}}
$$

(5 pts)

$$

(5 pts)

3x^2 - 12 = 0

(5 pts)

### **4. Expresiones Fraccionarias: Operaciones (20 puntos)**

a) Simplifique la fracción algebraica:

$$
\frac{x^2 - 4}{x^2 + 2x}
$$

(5 pts)

b) Efectúe la operación y simplifique:

$$
\frac{x}{x+2} + \frac{4}{x-2}
$$

(5 pts)

c) Multiplique y simplifique:

$$
\frac{x^2 - 9}{x^2 - x - 6} \cdot \frac{x^2 - 4}{x^2 - 3x}
$$

(5 pts)

d) Divida y simplifique:

$$
\frac{x^2 - 25}{x^2 + 6x + 5} \div \frac{x^2 + 3x + 2}{x^2 - 9}

(5 pts)

### **5. Desarrollo y Análisis Integrador (20 puntos)**

b) Determine los valores de $x$ para los cuales la expresión no está definida. (6 pts)
